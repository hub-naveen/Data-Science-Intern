# **Data Exploration and Preprocessing** 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset .csv')
print("Shapes of dataset:", df.shape) #no. of rows & columns

print("\nMissing values \n", df.isnull().sum()) # for missing values
```

# **Cuisines 9 has missing values**
## Fill missing 'Cuisines' with mode

```python
cuisine_mode = df['Cuisines'].mode()[0]
df['Cuisines'] = df['Cuisines'].fillna(cuisine_mode)
print(f"\n Filled missing 'Cuisines' with mode: {cuisine_mode}")
```

## Analyze distribution of target variables

```python
print("\nTarget variable distribution: \n", df['Aggregate rating'].value_counts())
```

# Plot

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'Aggregate rating', hue='Aggregate rating', data = df, palette='viridis', legend=False)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate rating')
plt.ylabel('Count')
plt.show()
```

## identify any class imbalances.

```python
class_counts = df['Aggregate rating'].value_counts()
imbalance_ratio = class_counts.max() / class_counts.min()
print(f"\n Imbalance ratio (max/min): {imbalance_ratio:.2f}")
```
